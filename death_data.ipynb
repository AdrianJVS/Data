```python
import pandas as pd
file = '/content/drive/MyDrive/DataSets/death_data.csv'
df = pd.read_csv(file)
df.head()
```

```python
df=df.fillna(0)
df=df.replace(':',0)
```

```python
df=df.drop(columns=['Flag and Footnotes','UNIT'])
```

```python
df=df.replace('Germany (until 1990 former territory of the FRG)','Germany')
```

```python
df['GEO'].unique()
```

```python
df['Year']=df['TIME'].str[0:4]
```

```python
df=df[(df['GEO']=='France') & (df['AGE']=='Total') & (df['SEX']=='Total')]
df.Value.replace(',','',inplace=True,regex=True)
df.Value=df.Value.astype('int')
df.pivot_table(index='Year', values='Value', aggfunc=sum)
```

```python
import plotly.express as px
fig = px.bar(df, x='Year', y='Value')
fig.show()
```
