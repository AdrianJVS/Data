```python
import pandas as pd
import numpy as np
file = '/content/drive/MyDrive/DataSets/births.xlsx'
df = pd.read_excel(file, sheet_name=2, header=8, skiprows=np.r_[9:15])
df.drop([col for col in df.columns if 'Unnamed:' in col],axis=1,inplace=True)
df = df.rename(columns= {'TIME':'Country'})
df = df.drop(labels=[29,30,31,54,55,56], axis=0)
```

```python
df = df.drop(df.index[range(36,51)])
df = df.drop(df.index[11])
df = df.drop(df.index[5])
```

```python
df=df.replace('Germany (until 1990 former territory of the FRG)','Germany')
df.drop(columns=['2019','2020'],inplace=True)
```

```python
df=df[df['Country']==['Poland','Germany']]
df.set_index('Country', inplace=True)
```

```python
df.T.plot(kind='bar')
```
