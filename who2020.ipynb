```python
import pandas as pd
import regex as re

file = '/content/drive/MyDrive/DataSets/who_2020.xlsx'
df = pd.read_excel(file, sheet_name='Annex 2-1', header=0)
pd.set_option('display.max_colwidth', None)
df.drop(index=206, inplace=True)
df.drop(columns=['Unnamed: 12', 'Unnamed: 16',	'Unnamed: 17',	'Unnamed: 18',	'Unnamed: 19',	'Unnamed: 20'], inplace=True)
df=df.rename(columns={'Unnamed: 0':'Country',
                   'Unnamed: 1':'Total populationa (000s) - Male', 'Unnamed: 2':'Total populationa (000s) - Female', 'Unnamed: 3':'Total populationa (000s) - Both Sexes',
                   'Unnamed: 4':'Life expectancy at birth (years) - Male', 'Unnamed: 5':'Life expectancy at birth (years) - Female', 'Unnamed: 6':'Life expectancy at birth (years) - Both Sexes',
                   'Unnamed: 7':'Healthy life expectancy at birth (years) - Male', 'Unnamed: 8':'Healthy life expectancy at birth (years) - Female', 'Unnamed: 9':'Healthy life expectancy at birth (years) - Both Sexes',
                   3.1:'Maternal mortality ratiod (per 100 000 live births)', 'Unnamed: 11':'Proportion of births attended by skilled health workers',
                   3.2:'Under-five mortality ratef (per 1000 live births)', 'Unnamed: 14':'Neonatal mortality ratef (per 1000 live births)', 3.3:'New HIV infectionsg (per 1000 uninfected population)'
                   })
df.drop(index=[0,1,2,3],inplace=True)
df.sort_values(by='Total populationa (000s) - Both Sexes', ascending=False, inplace=True)
df[df['Country'].isna()]
df.drop(index=198, inplace=True)
```
