```python
import pandas as pd
import numpy as np
file = '/content/drive/MyDrive/DataSets/births.xlsx'
df = pd.read_excel(file, sheet_name='Sheet 1', header=8, skiprows=np.r_[9:15])
df.drop([col for col in df.columns if 'Unnamed:' in col], axis = 1, inplace=True)
df = df.rename(columns={'TIME':'Country'})
df = df.drop(index=[5, 11])
df = df.replace('Germany (until 1990 former territory of the FRG)', 'Germany')
```

```python
df = df[(df['Country']=='Germany') | (df['Country']=='Poland')]
df=df.drop(columns='2020')
df
```

```python
df.set_index('Country', inplace=True)
df
```

```python
df.T.plot()
```

```python
dft=df.transpose()
dft['YoY Germany'] = dft['Germany'].pct_change().mul(100).round(2).astype(str)+'%'
dft['YoY Poland'] = dft['Poland'].pct_change().mul(100).round(2).astype(str)+'%'
dft.replace('nan%','', inplace=True)
dft['Year'] = dft.index
dft
```

```python
dft.columns
dft.to_dict()
```

```python
import plotly.express as px
year = tuple(list(dft['Year']))
births_germany = tuple(list(dft['Germany']))
births_poland = tuple(list(dft['Poland']))
fig = px.line(x=year, y=births_germany,
              labels=dict(x="Year", y="Births"),
              text=births_germany)
fig.update_traces(textposition='top center')
fig.add_bar(x=year, y=births_poland, name='Poland', text=births_poland)
fig.show()
```
